### Presentation of the result of the Pipeline
This jupyter notebook book is used to describe and present the result of the computed Mutual information of a music corpus. The mutual information has been computed over 3 midi files. 

```python
from src.reader.midi_reader import MidiReader
from src.reader.music_corpus import MusicCorpus
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt
from src.visualizer.plots import MutualInformationPlot
import numpy as np
import os
```

```python
CORPUS_LEN = 10
directory = './input/'

# allow to generate random music pieces
def generate_random_pieces(corpus_len,thres):
    x = np.random.rand(corpus_len,12)
    x[x>thres] = 1
    x[x<=thres] = 0
    return x
```

```python
mc = MusicCorpus(CORPUS_LEN)
for filename in os.listdir(directory):
    if filename.endswith(".mid"): 
        mid = MidiReader()
        mid.read(os.path.join(directory, filename))
        mid_arr = mid.get_np_array()

        mc.add_piece(mid_arr)
        
#add randomly generated pieces
'''   
for i in range(0,10):
        mid = MidiReader()
        mid_arr = generate_random_pieces(12,0.8)
        mc.add_piece(mid_arr)
    
'''
    

pcd = mc.compute_pitch_class_distr()
jpcd = mc.compute_join_pitch_class_distr()

mutual_info = mc.compute_mutual_info()        

plots = MutualInformationPlot(pcd.get_prob_density(),jpcd.get_prob_density(),mutual_info)
```

```python
mc.get_nbr_piece()
```

### Pitch class distribution
A matrix of size (CORPUS_LEN, NBR_Pitch)is the output of the computation of pitch class distribution. One pitch distribution for each time instant can be extracted.


```python
plots.pitch_class_distr_overall_plot()
```

### Join Pitch class distribution
A matrix of size (CORPUS_LEN, CORPUS_LEN, NBR_PITCH, NBR_PITCH) encodes all the informations to extract the join pitch class distribution parametrize by two time instant. Here the 2d Histrogram plot of the join pitch class density at the discrete time intants  t = 0 and t = 8


```python
plots.join_distr_plot(0,9)
```

### Mutual information of the Music Corpus
Using the Pitch class distribution, Join pitch class distribution, we can compute the Mutual Information present in the music corpus. A matrix (CORPUS_LEN, CORPUS_LEN) encodes all the mutual information of any time instants.

```python
plots.mutual_info_overall_plot()
```
