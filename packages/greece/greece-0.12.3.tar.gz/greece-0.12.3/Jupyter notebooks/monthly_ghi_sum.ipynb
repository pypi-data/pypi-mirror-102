# Compute monthly sums of GHI

```python
from gis_tools.raster import RasterMap
from utils.sys.browser import find_file
```

Iterate over GHI raster maps in folder:

```python
old_path = "/home/benjamin/Documents/Data/Resource rasters/Solar maps/Monthly GHI/"
new_path = "/home/benjamin/Documents/Data/Resource rasters/Solar maps/Monthly GHI sum/"
solar_map_files = find_file(".tif", old_path)
solar_map_files =sorted(solar_map_files)
day_num = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
for n, file in enumerate(solar_map_files):
    smap = RasterMap(file, no_data_value=-9999)
    smap = smap * day_num[n]
    smap.to_file(new_path + 'monthly_GHI_%02d.tif' % (n+1))
```

```python
april_map = RasterMap(new_path + 'monthly_GHI_04.tif', no_data_value=-9999)
april_map.plot(colorbar=True, colorbar_title="Monthly solar irradiation (Wh/mÂ²)")
```

```python
april_map = RasterMap(old_path + 'monthly_GHI_04.tif', no_data_value=-9999)
april_map.plot(colorbar=True, colorbar_title="Daily solar irradiation (Wh/mÂ².day)")
```
